```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,301)
u = 0.1 * x**2
p = 0.5*np.exp(-u)
plt.plot(x,u,label="U(x)")
plt.plot(x,p,":",label="p(x)=exp(-U(x)/kT)")
plt.legend()
```

```python
unl = -0.5 * x**2 + 0.1*x**4 + 0.7
pnl = 3*np.exp(-unl)
plt.plot(x,unl,label="U(x)")
plt.plot(x,pnl,":",label="p(x)=exp(-U(x)/kT)")
plt.legend()
```
