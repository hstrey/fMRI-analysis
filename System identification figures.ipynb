```python
import numpy as np
import scipy.stats as sps
import os
import csv
import matplotlib.pyplot as plt
from scipy.stats import skewtest, skew, kurtosis, kurtosistest
from scipy.optimize import root

SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

SEED = 350107321 # from random.org
np.random.seed(SEED)

print(plt.style.available)
plt.style.use('seaborn-white')
```

```python
t = np.linspace(0,1,100)
y = np.exp(-t*3)
plt.plot(t,y,color="k",linewidth=8)
plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False, # labels along the bottom edge are off
    labelleft=False) 
plt.savefig("expdecay.png",format='png',dpi=300,bbox_inches='tight',facecolor="white",backgroundcolor="white")
```

```python
t = np.linspace(0,1,100)
y = np.exp(t)
plt.plot(t,y,color="k",linewidth=5)
plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False, # labels along the bottom edge are off
    labelleft=False) 
plt.savefig("expgrowth.png",format='png',dpi=300,bbox_inches='tight',facecolor="white",backgroundcolor="white")
```

```python
y = np.exp(-t*3)*np.cos(t*3*np.pi)
plt.plot(t,y,color="k",linewidth=8)
plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False, # labels along the bottom edge are off
    labelleft=False) 
plt.axhline(y=0, color='k', linestyle='-')
plt.savefig("expdecayosc.png",format='png',dpi=300,bbox_inches='tight',facecolor="white",backgroundcolor="white")
```
